```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
#Loading the dataset
# The dataset is a CSV file containing information about red wine quality.
# The dataset is LOADED using the pandas library
# Copied full path of the file for other ways brought up errors

wine_data = pd.read_csv(r'C:\Users\homel\WeekAss_2\ds_env\Scripts\winequality-red.csv', sep=';')
```

```python
# Loading the first 10 rows of the dataset

wine_data.head(10)
```

```python
#Renaming the columns
# The columns are renamed to remove spaces and make them more readable.

labels = list(wine_data.columns)
labels[0] = labels[0].replace(' ', '_')
labels[1] = labels[1].replace(' ', '_')
labels[2] = labels[2].replace(' ', '_')
labels[3] = labels[3].replace(' ', '_')
labels[4] = labels[4].replace(' ', '_')
labels[5] = labels[5].replace(' ', '_')
labels[6] = labels[6].replace(' ', '_')

wine_data.columns = labels

# Displaying the first 5 rows of the dataset after renaming the columns
wine_data.head()

```

```python
# Exploring the dataset

print('\n DATA INFORMATION:')
wine_data.info()

print('\n DATA STATISTICS:')
wine_data.describe()


```

```python
# cHECKING FOR NULL VALUES
# Checking for null values isnull() method.

wine_data.isnull()

# Using sum() method to count the number of null values in each column.

wine_data.isnull().sum()

#FINDINGS
# The dataset does not contain any null values, as the sum of null values in each column is 0.

```

```python
# Data Cleaning
# Checking for duplicates using the duplicated() method and the sum() method.

duplicates = wine_data.duplicated().sum()
print(f'\n Number Of Duplicates:, {duplicates}')

wine_data[wine_data.duplicated()]

```

```python
#Removing Duplicates
# Creating a new DataFrame without duplicates using the drop_duplicates() method.
# The original DataFrame is not modified, and the new DataFrame is stored in a variable called wine_data2.

wine_data2 = wine_data.drop_duplicates()
print(f"Number of Duplicates After Cleaning: ", wine_data2.duplicated().sum())
```

```python
#Drop Columns that are not needed for analysis
# The quality in the data is the dependant variable thus cant be used for analysis
# Creating a new DataFrame without the 'quality' column using the drop() method.
# The original DataFrame is not modified, and the new DataFrame is stored in a variable called wine_data3.

wine_data3 = wine_data2.drop(['quality'], axis=1)
```

```python
wine_data3.head()
```

```python
#Visualization of the dataset for no good reason
#Finding out there different colormap names used in palette
# Added hue to the countplot to show the distribution of quality

plt.figure(figsize=(15, 10))
sns.countplot(x='quality', data=wine_data, palette='cividis', hue='quality')
plt.xlabel('Quality')
plt.ylabel('Count')

plt.title('Distribution of Wine Quality')        
plt.show()


# Distribution of features
# Visualizing the distribution of each feature using histograms

features = wine_data3.columns # Excluded the 'quality' column   
plt.figure(figsize=(20, 15))
for i, feature in enumerate(features):
    plt.subplot(3, 4, i + 1)
    sns.histplot(wine_data3[feature], bins=20, kde=True, color='blue')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(f'Distribution of {feature}')
plt.tight_layout()
plt.show()

```

```python
# Removing outliers
# Checking for outliers using the z-score method.
from scipy import stats

z_scores = np.abs(stats.zscore(wine_data3)) #Wine_data3 is the DataFrame without the duplicates and 'quality' column
threshold = 3
outliers = np.where(z_scores > threshold)
print(f'Number of outliers: {len(outliers[0])}')



```

```python
#Removing outliers from our wine_data3 DataFrame
# Creating a new DataFrame without outliers using the z-score method.   
# The original DataFrame is not modified, and the new DataFrame is stored in a variable called wine_data4.
wine_data4 = wine_data3[(z_scores < threshold).all(axis=1)]




```
